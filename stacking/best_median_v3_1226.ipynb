```python
import pickle
# https://www.kaggle.com/submarineering/submarineering-even-better-public-score-until-now
ens_file = 'all_xgb_sub_fold_3_rnd_42_1244.csv' # 1244
other_files = [
    '../results/cnn_1_aug_1_fold8.csv',
    '../results/cnn_2_aug_1_fold8.csv',
    '../results/cnn_3_aug_1_fold8.csv',
    '../results/cnn_5_aug_1_fold8.csv',
    '../results/cnn_4_aug1_best_acc_sub.csv',
    '../results/incept_aug1_sub.csv',
    '../results/resnet_aug1_sub.csv',
    '../results/xception_aug1_sub.csv',
    '../results/vgg_aug1_sub.csv',
    '../features/other_model_gbrt2.pkl',
    '../features/other_model_lgb2.pkl',
    '../features/other_model_lr2.pkl',
    '../features/other_model_rf2.pkl',
    '../features/other_model_xgb2.pkl',
]

import pandas as pd
best_df = pd.read_csv(ens_file)
best_df.head()
```

```python
other_res = []
for f in other_files:
    if f.endswith('csv'):
        other_res.append(pd.read_csv(f).is_iceberg.values)
    else:
        with open(f,'rb') as fin:
            a,b = pickle.load(fin)
            other_res.append(b.flatten())
    
import numpy as np
other_res = np.vstack(other_res).T
```

```python
concat_df = pd.DataFrame(other_res,columns=['a{}'.format(i) for i in range(len(other_files))])
concat_df['base'] = best_df.is_iceberg
concat_df.head()
```

```python
concat_sub = concat_df.copy()
concat_sub['is_iceberg_base'] = best_df['is_iceberg']
all_files_cnt = len(other_files)+1
concat_sub['is_iceberg_max'] = concat_sub.iloc[:, 0:all_files_cnt].max(axis=1)
concat_sub['is_iceberg_min'] = concat_sub.iloc[:, 0:all_files_cnt].min(axis=1)
concat_sub['is_iceberg_mean'] = concat_sub.iloc[:, 0:all_files_cnt].mean(axis=1)
concat_sub['is_iceberg_median'] = concat_sub.iloc[:, 0:all_files_cnt].median(axis=1)

concat_sub.head()
```

```python
cutoff_lo = 0.8
cutoff_hi = 0.2
concat_sub['is_iceberg'] = np.where(np.all(concat_sub.iloc[:,0:all_files_cnt] > cutoff_lo, axis=1), 
                                    concat_sub['is_iceberg_max'], 
                                    np.where(np.all(concat_sub.iloc[:,0:all_files_cnt] < cutoff_hi, axis=1),
                                             concat_sub['is_iceberg_min'], 
                                             concat_sub['is_iceberg_base']))
#concat_sub['is_iceberg'] = np.clip(concat_sub['is_iceberg'].values, 0.001, 0.999)
concat_sub['id'] = best_df['id']
concat_sub[['id', 'is_iceberg']].to_csv('stack_minmax_bestbase3.csv', 
                                        index=False)

concat_sub[['id', 'is_iceberg']].head(10)
```

```python
diff_res = concat_sub['is_iceberg']-concat_sub['base']
print(type(diff_res))
cnt = np.where(diff_res!=0)
print(cnt[0].shape)
print(concat_sub['is_iceberg'].std())
```
